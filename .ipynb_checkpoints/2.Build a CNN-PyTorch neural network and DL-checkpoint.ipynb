# 在pytorch中建立一个神经网络
1. 创建一个神经网络类扩展模型基类 nn.Module
2. 在类构造函数中将网络的层定义为类属性
3. 使用网络层属性以及nn.functonal API的操作   

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Network(nn.Module):
    def __init__(self):
        super(Network, self).__init__()
        #两个卷积层
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
        #kernel_size: 设置该层中滤波器大小
        #out_channels: 设置滤波器数量
        #在一个卷积层内,输入通道与一个卷积滤波器配对执行卷积运算,操作的结果是输出通道, 每个卷积滤波器对应一个输出通道,所以out_channels是卷积滤波器数量
        #第一个卷积层的in_channels依赖于训练集图像的彩色通道数量
        
        
        #三个线性层
        self.fc1 = nn.Linear(in_features=12*4*4, out_features=120)
        self.fc2 = nn.Linear(in_features=120, out_features=60)
        self.out = nn.Linear(in_features=60, out_features=10)
        #out_features 输出张量大小
        #最后一个线性层的out_features依赖于训练集中类的数量
        #一层的输入是上一层的输出
    def forward(self,t):
        return t
```

## 实现网络前向方法

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Network(nn.Module):
    def __init__(self):
        super(Network,self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
        
        self.fc1 = nn.Linear(in_features=12*4*4,out_features=120)
        self.fc2 = nn.Linear(in_features=120, out_features=60)
        out = nn.Linear(in_features=60, out_features=10)
        
    def forward(self, t):
        #1 输入层
        t = t
        
        #2 隐藏卷积层
        t = self.conv1(t) #卷积操作
        t = F.relu(t) #激活操作
        t = F.max_pool2d(t, kernel_size = 2, stride = 2) #最大池化操作
        
        #3 隐藏卷积层
        t = self.conv2(t)
        t = F.relu(t)
        t = F.max_pool2d(t, kernel_size = 2, stride = 2) 
        
        #4 隐藏线性层
        t = t.reshape(-1, 12*4*4)
        t = self.fc1(t)
        t = F.relu(t)
        
        #5 隐藏线性层
        t = self.fc2(t)
        t = F.relu(t)
        
        #6 输出层
        t = self.out(t)
        #t = F.softmax(t,dim = 1)
        return t
```

```python
#网络类
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
            
        self.fc1 = nn.Linear(in_features=12*4*4, out_features=120)
        self.fc2 = nn.Linear(in_features=120, out_features=60)
        self.out = nn.Linear(in_features=60, out_features=10)
            
    def forward(self,t):
        t = F.relu(self.conv1(t))
        t = F.max_pool2d(t, kernel_size = 2, stride = 2)
        
        t = F.relu(self.conv2(t))
        t = F.max_pool2d(t, kernel_size = 2, stride = 2)
        
        t = F.elu(self.fc1(t.reshape(-1, 12*4*4)))
        t = F.relu(self.fc2(t))
 
        t = self.out(t)
        
        return t
```

```python
#定义一个get_num_correct函数,接受预测张量和标签张量,使用item返回预测正确个数
def get_num_correct(preds, labels):
    return preds.argmax(dim = 1).eq(labels).sum().item()

# preds.argmax(dim = 1) 得到预测最大可能行标签
# preds.argmax(dim=1).eq(labels) 与labels进行比较得到每个预测正确结果
# preds.argmax(dim=1).eq(labels).sum() 得到预测正确数量
```

# 训练过程
1. 从训练集中得到批数据
2. 把批数据传入网络中
3. 计算损失函数
4. 计算损失函数的梯度和网络权值
5. 更新权重,用梯度减少损失
6. 重复1-5,直到一个周期完成
7. 重复1-6,获得期望的精确度

# 训练一个pytorch CNN、计算loss、梯度和更新权重

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import torchvision
import torchvision.transforms as transforms

torch.set_printoptions(linewidth=120)
torch.set_grad_enabled(True)
```

```python
def get_num_correct(preds, labels):
    return preds.argmax(dim=1).eq(labels).sum().item()
```

```python
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=12, kernel_size=5)
        
        self.fc1 = nn.Linear(in_features=12*4*4, out_features=120)
        self.fc2 = nn.Linear(in_features=120, out_features=60)
        self.out = nn.Linear(in_features=60, out_features=10)
        
    def forward(self, t):
        t = F.relu(self.conv1(t))
        t = F.max_pool2d(t, kernel_size = 2, stride = 2)
        
        t = F.relu(self.conv2(t))
        t = F.max_pool2d(t, kernel_size = 2, stride = 2)
        
        t = t.reshape(-1,12 * 4 *4)
        t = F.relu(self.fc1(t))
        
        t = F.relu(self.fc2(t))
        
        t = self.out(t)
        
        return t
```

```python
network = Network()
```

```python
train_set = torchvision.datasets.FashionMNIST(
    root= './data/FasionMINIST/',
    train=True,
    download=True,
    transform = transforms.Compose([
        transforms.ToTensor()
    ])
)
```

```python
train_loader = torch.utils.data.DataLoader(train_set, batch_size=100)
batch = next(iter(train_loader))
images, labels = batch
```

# CNN训练循环

## 训练所有的batch:一个epoch

```python
network = Network()

train_loader = torch.utils.data.DataLoader(train_set, batch_size=100)
optimizer = optim.Adam(network.parameters(), lr = 0.01)

total_loss = 0
total_correct = 0

for batch in train_loader:
    images, labels = batch
    preds = network(images)
    loss = F.cross_entropy(preds, labels)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    total_loss += loss.item()
    total_correct += get_num_correct(preds, labels)
    
print('epoch:', 0, 'total correct:', total_correct, 'loss: ',total_loss)
```

```python
#准确率
total_correct / len(train_set)
```

# 训练多次epoch

```python
network = Network()

train_loader = torch.utils.data.DataLoader(train_set, batch_size=100)
optimizer = optim.Adam(network.parameters(), lr = 0.01)
for epoch in range(5):
    total_loss = 0
    total_correct = 0

    for batch in train_loader:
        images, labels = batch
        preds = network(images)
        loss = F.cross_entropy(preds, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        total_correct += get_num_correct(preds, labels)

    print('epoch:', epoch, 'total correct:', total_correct, 'loss: ',total_loss,'correcting rate:', total_correct/len(train_set))
```

# 构建混淆矩阵

## 得到整个训练集的预测结果

```python
def get_all_preds(model, loader):
    all_preds = torch.tensor([])
    for batch in loader:
        images, labels = batch
        preds = model(images)
        
        all_preds = torch.cat(
            (all_preds, preds),
            dim = 0
        )
    return all_preds
```

```python
prediction_loader = torch.utils.data.DataLoader(train_set, batch_size=10000)
train_preds = get_all_preds(network, prediction_loader)
```

```python
train_preds.shape
```

```python
print(train_preds.requires_grad) #预测张量的梯度属性
```

```python
# 做预测时,不希望跟踪梯度,或者不需要创建表,所以关闭,节省资源
with torch.no_grad():
    prediction_loader = torch.utils.data.DataLoader(train_set, batch_size=10000)
    train_preds = get_all_preds(network, prediction_loader)
```

```python
print(train_preds.requires_grad)# 没有设置梯度属性
```

```python
train_preds.grad
```

```python
preds_correct = get_num_correct(train_preds,train_set.targets)

print('total correct: ', preds_correct)
print('accuracy: ', preds_correct/len(train_set))
```
